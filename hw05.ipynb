```python
# 1. Генерація 100 випадкових значень ознак x1, x2
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(100)
X = np.random.rand(100, 2)
x1 = X[:, 0]
x2 = X[:, 1]

# Реалізація функції polynomial(x1, x2)
def polynomial(x1, x2):
    # Поліном: 4x1^2 + 5x2^2 - 2x1*x2 + 3x1 - 6x2
    return 4*x1**2 + 5*x2**2 - 2*x1*x2 + 3*x1 - 6*x2

# Обчислення цільової змінної y (розмірність 100,)
y = polynomial(x1, x2)
print(y)
```

```python
# 2. Генерація додаткових ознак за допомогою PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly.fit_transform(X)

print(f"Розмірність початкових даних X: {X.shape}")
print(f"Розмірність y: {y.shape}")
print(f"Розмірність матриці поліноміальних ознак X_poly: {X_poly.shape}")
```

```python
#3. Реалізація функції для методів градієнтного спуску.
# Функція polynomial_regression_gradient_descent() для обчислення коефіцієнтів поліноміальної регресії методом градієнтного спуску
def polynomial_regression_gradient_descent(X, y, learning_rate=0.1, n_iterations=1000):
    m = len(y)
    theta = np.zeros(X.shape[1])
    for _ in range(n_iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        gradients = (1/m) * X.T.dot(errors)
        theta -= learning_rate * gradients
    return theta

# Функція polynomial_regression_SGD() варіанта SGD метода градієнтного спуску
def polynomial_regression_SGD(X, y, learning_rate=0.01, n_iterations=50):
    m = len(y)
    theta = np.zeros(X.shape[1])
    for epoch in range(n_iterations):
        for i in range(m):
            random_index = np.random.randint(m)
            xi = X[random_index:random_index+1]
            yi = y[random_index:random_index+1]
            predictions = xi.dot(theta)
            errors = predictions - yi
            gradients = xi.T.dot(errors)
            theta -= learning_rate * gradients
    return theta

# Функція polynomial_regression_rmsprop() варіанта RMSProp метода градієнтного спуску.
def polynomial_regression_rmsprop(X, y, learning_rate=0.1, n_iterations=1000, decay_rate=0.9, eps=1e-8):
    m = len(y)
    theta = np.zeros(X.shape[1])
    cache = np.zeros(X.shape[1])
    for _ in range(n_iterations):
        gradients = (1/m) * X.T.dot(X.dot(theta) - y)
        cache = decay_rate * cache + (1 - decay_rate) * (gradients ** 2)
        theta -= learning_rate * gradients / (np.sqrt(cache) + eps)
    return theta

# Функція polynomial_regression_adam() варіанта Adam метода градієнтного спуску.
def polynomial_regression_adam(X, y, learning_rate=0.1, n_iterations=1000, beta1=0.9, beta2=0.999, eps=1e-8):
    m = len(y)
    theta = np.zeros(X.shape[1])
    m_t = np.zeros(X.shape[1])
    v_t = np.zeros(X.shape[1])
    for t in range(1, n_iterations + 1):
        gradients = (1/m) * X.T.dot(X.dot(theta) - y)
        m_t = beta1 * m_t + (1 - beta1) * gradients
        v_t = beta2 * v_t + (1 - beta2) * (gradients ** 2)
        m_hat = m_t / (1 - beta1**t)
        v_hat = v_t / (1 - beta2**t)
        theta -= learning_rate * m_hat / (np.sqrt(v_hat) + eps)
    return theta

# Функція polynomial_regression_nadam() варіанта Nadam метода градієнтного спуску.
def polynomial_regression_nadam(X, y, learning_rate=0.1, n_iterations=1000, beta1=0.9, beta2=0.999, eps=1e-8):
    m = len(y)
    theta = np.zeros(X.shape[1])
    m_t = np.zeros(X.shape[1])
    v_t = np.zeros(X.shape[1])
    for t in range(1, n_iterations + 1):
        gradients = (1/m) * X.T.dot(X.dot(theta) - y)
        m_t = beta1 * m_t + (1 - beta1) * gradients
        v_t = beta2 * v_t + (1 - beta2) * (gradients ** 2)
        m_hat = m_t / (1 - beta1**t)
        v_hat = v_t / (1 - beta2**t)

        m_hat_nesterov = beta1 * m_hat + (1 - beta1) * gradients / (1 - beta1**t)
        theta -= learning_rate * m_hat_nesterov / (np.sqrt(v_hat) + eps)
    return theta
```

```python
# 4.Базові налаштування ітерацій для досягнення збіжності
print("Перевірка знайдених коефіцієнтів (справжні: [0, 3, -6, 4, -2, 5]):\n")

theta_gd = polynomial_regression_gradient_descent(X_poly, y, n_iterations=2000)
print(f"GD: {np.round(theta_gd, 2)}")

theta_sgd = polynomial_regression_SGD(X_poly, y, n_iterations=200)
print(f"SGD: {np.round(theta_sgd, 2)}")

theta_rmsprop = polynomial_regression_rmsprop(X_poly, y, n_iterations=500)
print(f"RMSprop: {np.round(theta_rmsprop, 2)}")

theta_adam = polynomial_regression_adam(X_poly, y, n_iterations=500)
print(f"Adam: {np.round(theta_adam, 2)}")

theta_nadam = polynomial_regression_nadam(X_poly, y, n_iterations=500)
print(f"Nadam: {np.round(theta_nadam, 2)}")
```

* За часом виконання найефективнішим виявився RMSprop, тоді як SGD показав найгірший результат.
