```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

# Фіксуємо seed для відтворюваності результатів
np.random.seed(42)

# Крок 1.1: Генерація 100 випадкових значень ознак x1, x2 (масив 100x2)
X = np.random.rand(100, 2)
x1 = X[:, 0]
x2 = X[:, 1]

# Крок 1.2: Реалізація функції polynomial(x1, x2)
def polynomial(x1, x2):
    # Поліном: 4x1^2 + 5x2^2 - 2x1*x2 + 3x1 - 6x2
    return 4*x1**2 + 5*x2**2 - 2*x1*x2 + 3*x1 - 6*x2

# Обчислення цільової змінної y (розмірність 100,)
y = polynomial(x1, x2)

# Крок 2: Генерація додаткових ознак за допомогою PolynomialFeatures
# degree=2 створить ознаки: 1, x1, x2, x1^2, x1*x2, x2^2
poly = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly.fit_transform(X)

print(f"Розмірність початкових даних X: {X.shape}")
print(f"Розмірність y: {y.shape}")
print(f"Розмірність матриці поліноміальних ознак X_poly: {X_poly.shape}")
```
